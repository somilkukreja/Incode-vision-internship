```python
# adding lib
import pandas as pd
import numpy as np
```

```python
# loading dataset
df = pd.read_csv("/content/student-lifestyle-and-stress-dataset.csv")

print("Dataset Loaded Successfully!")
```

```python
# view dataset
df.head()
```

```python
# dataset information
df.info()
```

```python
# dataset shape
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# checking missing values
df.isnull().sum()
```

```python
# checking duplicate records
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# removing duplicate
df = df.drop_duplicates()

print("Shape After Removing Duplicates:")
print(df.shape)
```

```python
# identifying numerical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

print(num_cols)
```

```python
# filling missing values
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

print("Numerical Missing Values Handled")
```

```python
# Identifyinc catogerical columns
cat_cols = df.select_dtypes(include=['object']).columns

print(cat_cols)
```

```python
# filling missing catagorical values
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Categorical Missing Values Handled")
```

```python
# verifying missing values
df.isnull().sum()
```

```python
#standardize text columns
for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

print("Text Formatting Completed")
```

```python
# checking data type
df.dtypes
```

```python
# Statistical Summary
df.describe()
```

```python
# detect and handel outlaires
for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - (1.5 * IQR)
    upper = Q3 + (1.5 * IQR)

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outliers Handled")
```

```python
# encoding catagorical variables
df_clean = pd.get_dummies(df, drop_first=True)

print("New Shape:", df_clean.shape)
```

```python
# feature scqaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
# save cleaned dataset
df_clean.to_csv(
    "cleaned_student_lifestyle_stress_dataset.csv",
    index=False
)

print("Cleaned Dataset Saved Successfully")
```
